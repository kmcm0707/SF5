```python
poission_graph = poisson_configuration_graph(10000, 10)
#print(poission_graph.number_of_triangles_quick())
triangle_poisson_graph = copy.deepcopy(poission_graph)
edges_added = 0
for i in range(50):
    edges_added += triangle_poisson_graph.add_triangle_2()
print(edges_added)
standeredised_poisson_graph = copy.deepcopy(poission_graph)
for i in range(edges_added):
    standeredised_poisson_graph.add_random_edge()
```

```python
quick_recovery = []
quick_recovery_tri = []
for i in range(50):
    temp = [SIR_simulation(standeredised_poisson_graph, 0.01*i, 3, 400, 0.2)[2][-1] for _ in range(20)]
    quick_recovery.append(np.mean(temp))

for i in range(50):        
    temp_tri = [SIR_simulation(triangle_poisson_graph, 0.01*i, 3, 400, 0.2)[2][-1] for _ in range(20)]  
    quick_recovery_tri.append(np.mean(temp_tri))

plt.plot([0.01*i for i in range(50)], quick_recovery, label = "Standardised Poisson", color = "blue")
plt.plot([0.01*i for i in range(50)], quick_recovery_tri, label = "Triangle Poisson", color = "red")
plt.xlabel("Infection Probability")
plt.ylabel("People Recovered")
plt.legend()
plt.savefig("triangle_poisson_vs_standardised_poisson_vaccine.png")
plt.show()

```
